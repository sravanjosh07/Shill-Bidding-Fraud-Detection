```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.model_selection import train_test_split
```

```python
# Load data from CSV file
data = pd.read_csv("Group_14_data_cleaned.csv")

# Split the data into training and testing sets
X = data.drop(['Class', 'Bidder_ID'], axis=1)
y = data['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state= 26)
```

```python
inertia = []

for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit(X_train)
    inertia.append(km.inertia_)

    
```

```python
inertia
plt.plot(range(1,16), inertia)
plt.xlabel('K', fontsize=15)
plt.ylabel('inertia', fontsize=15);
```

From the above plot, we can see that the optimal value for k is 3 (using the elbow method)

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for k in range(2,6):

    km = KMeans(n_clusters=k)
    q, mod = divmod(k, 2)

    visualizer = SilhouetteVisualizer(km, ax=ax[q-1][mod])
    visualizer.fit(X)
```

```python
# Fit k-means algorithm to the data
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


# Visualize the clusters using scatter plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clustering using k-means algorithm')
plt.show()
```
