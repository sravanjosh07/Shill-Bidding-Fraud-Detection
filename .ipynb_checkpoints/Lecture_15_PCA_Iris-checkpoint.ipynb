# Principal Component Analysis (PCA) 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn import preprocessing     
````

````python
df = pd.read_csv('Group_14_data_cleaned.csv') 
df.head()
````

````python
#Dropping the target variable and two features as they are of no significance
X = df.drop(['Bidder_ID','Auction_ID', 'Class'],axis = 1)
````

````python
#Scaling the data
X_scaled = preprocessing.scale(X) 
````

````python
pd.DataFrame(data = X_scaled) 
````

````python
pca = PCA()    

pca.fit(X_scaled)
````

````python
per_var = np.round(pca.explained_variance_ratio_*100, 2)
per_var
````

````python
princ_comp = ['PC' + str(x) for x in range(1, 10)]

plt.bar(x = range(1,10), height = per_var, tick_label = princ_comp)
plt.xlabel('Principal Component', fontsize = 15)
plt.ylabel('Percentage of Variation', fontsize = 15)
plt.title('Scree Plot', fontsize = 15)
plt.grid();
````

````python
pd.DataFrame(data = pca.components_, columns = princ_comp)
````

----

We use ```pca.transform``` to see how the projected data looks like in the new coordinate system

This will be used later to plot the projected data in the first two coordinates PC1, PC2

---

````python
pca_data = pca.transform(X_scaled)

pca_df = pd.DataFrame(pca_data, columns = princ_comp)

pca_df
````

````python
# We delete PC3 to PC9 because we decided to use only the first two components

del pca_df['PC4'], pca_df['PC5'], pca_df['PC6'], pca_df['PC7'], pca_df['PC8'], pca_df['PC9']

pca_df
````

````python
# We add to the previous data frame the labels 

projected_data = pd.concat([pca_df, df[['Class']]], axis = 1)

projected_data
````

---

We plot the data points in this 2-dimensional coordinates system given by PC1 and PC2 

We also color each data point according to its label

---

````python
labels = [0,1]
colors = ["r", "b"]

for label, color in zip(labels, colors):
    index = projected_data["Class"]==label
    plt.scatter(projected_data.loc[index, "PC1"],
                projected_data.loc[index, "PC2"],
                c = color)
    
plt.xlabel("PC1", fontsize = 15)
plt.ylabel("PC2", fontsize = 15)
plt.title("Two-component PCA", fontsize = 15)
plt.legend(labels);
````

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


x = pca_df['PC1']
y = pca_df['PC2']
z = pca_df['PC3']
colors = ["r","b", "g"]

# Create a figure and a 3D axes object
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(x, y, z, color = "r" if(), marker='o')

# Set the labels for the axes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
````
